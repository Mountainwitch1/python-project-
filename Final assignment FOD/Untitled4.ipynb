```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("cities.csv")
print(data.columns)
```

```python
# Step 2: Calculate the total landmass area covered by each continent
continent_landmass = data.groupby('continent')['area'].sum()

# Step 3: Create a bar plot
continent_landmass.plot(kind='bar')

# Set plot labels and title
plt.xlabel('Continent')#yo x axis ho
plt.ylabel('Landmass Area')#yo y axis ho
plt.title('Landmass Area Covered by Each Continent')# Title for the plot

# Display the plot
plt.show()
```

```python
# b. Cities with highest and lowest elevations for each continent
continents = data['continent'].unique()

for continent in continents:
    continent_data = data[data['continent'] == continent]
    
    # City with highest elevation
    highest_elevation_city = continent_data[continent_data['elevation'] == continent_data['elevation'].max()]
    highest_elevation = highest_elevation_city['elevation'].values[0]#high elev value
    highest_lat = highest_elevation_city['lat'].values[0]#lat of high ele
    highest_lon = highest_elevation_city['long'].values[0]#long ho high elev 
    
    # City with lowest elevation
    lowest_elevation_city = continent_data[continent_data['elevation'] == continent_data['elevation'].min()]
    lowest_elevation = lowest_elevation_city['elevation'].values[0]#low elevation val
    lowest_lat = lowest_elevation_city['lat'].values[0]#low elevation lat
    lowest_lon = lowest_elevation_city['long'].values[0]# longitude of the lowest city
    
    # Print the results
    print(f'Continent: {continent}')
    print(f'Highest Elevation: City = {highest_elevation_city["country"].values[0]}, Elevation = {highest_elevation}, Latitude = {highest_lat}, Longitude = {highest_lon}')
    print(f'Lowest Elevation: City = {lowest_elevation_city["country"].values[0]}, Elevation = {lowest_elevation}, Latitude = {lowest_lat}, Longitude = {lowest_lon}')
    print('---')


```

```python
# Task c: Predict cities with highest and lowest population density
import pandas as pd

# Step 1: Load the data from "cities.csv" file
data = pd.read_csv("cities.csv")

# Step 2: Calculate the population density for each city
data['population'] = data['population'] / data['area']

# Step 3: Sort the cities based on population density in ascending and descending order
sorted_cities_lowest_density = data.sort_values('population').head(5)
sorted_cities_highest_density = data.sort_values('population', ascending=False).head(5)

# Step 4: Retrieve the top five cities with the highest and lowest population density
cities_lowest_density = sorted_cities_lowest_density['country'].tolist()
cities_highest_density = sorted_cities_highest_density['country'].tolist()

# Print the results
print("Cities with the lowest population density:")
for city in cities_lowest_density:
    print(city)

print("\nCities with the highest population density:")
for city in cities_highest_density:
    print(city)
```
